```python
import numpy as np
import utils
import models.regression as regmodels
import models.classification as classmodels
from sklearn.metrics import mean_squared_error
import torch
import pandas as pd
import torch.nn as nn
```

```python
from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import LabelEncoder

# fetch dataset
chess_king_rook_vs_king_pawn = fetch_ucirepo(id=22)

# data (as pandas dataframes)
X = chess_king_rook_vs_king_pawn.data.features
y = chess_king_rook_vs_king_pawn.data.targets

for column in X.columns:
    if X[column].dtype == 'object':
        le = LabelEncoder()
        X[column] = le.fit_transform(X[column])

y = LabelEncoder().fit_transform(y)
X
```

```python
model = classmodels.AnnLassoClassifierl0()
model.fit(X, y, 1)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(model.predict(X), y)
```

```python
X = np.random.normal(size=(100, 200))
s = 4
features = np.arange(s)
b = 3*np.ones(s)
y = np.dot(X[:, features], b) + np.random.normal(size=X.shape[0])

model = regmodels.AnnLassoRegressorl0(hidden_dims=(20, 10))
model.fit(X,y,1)
print(mean_squared_error(y, model.predict(X)))
print(model.imp_feat)
```
